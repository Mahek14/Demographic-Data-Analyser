```python
import pandas as pd
```

```python
df = pd.read_csv("adult.data.csv")
```

```python
df.shape
```

```python
race_count = df.groupby('race')['race'].count().sort_values(ascending=False)
race_count
```

```python
average_age_men =round( df.groupby('sex', as_index=True).age.mean()['Male'],1)
average_age_men
```

```python
a= round(df[df['education'] == 'Bachelors'].shape[0] / df.shape[0] * 100,1)
a
```

```python
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
```

```python
higher_education_rich = round(higher_education[higher_education['salary'] == '>50K']['salary'].count() / higher_education.shape[0] * 100,1)
print(higher_education_rich,'%')
```

```python
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
lower_education
```

```python
lower_education_rich =  round(lower_education[lower_education['salary'] == '>50K']['salary'].count() / lower_education.shape[0] * 100,1)
lower_education_rich
```

```python
print(round(df['hours-per-week'].min(),1))
```

```python
highest_earning_country = (df[df['salary'] == '>50K']['native-country'].value_counts()/df['native-country'].value_counts() * 100).idxmax()
print(highest_earning_country)
```

```python
top_IN_occupation = df[(df['salary'] == '>50K') & (df['native-country'] == 'India')].groupby('occupation')['occupation'].count().idxmax()
print(top_IN_occupation)
```

```python
highest_earning_country_percentage = df[((df['native-country'] == highest_earning_country)& (df['salary']== '>50K') )]
highest_earning_country_percentage
```
